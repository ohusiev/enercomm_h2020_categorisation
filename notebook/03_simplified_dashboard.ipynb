```python
import pandas as pd
df= pd.read_csv('../data/matrix_visualisation.csv', index_col='Unnamed: 0')
#df["Main_categoty"] = df["Main_categoty"].astype(str)
```

```python
classifier_subcateg = pd.read_csv('../data/classifier_types.csv', index_col='id')
subcat_dict={}
for index, row in classifier_subcateg.iterrows():
    if index>10:
        subcat_dict[row['Cathegories']] = str(index) 

subcat_dict['Keywords match - Total ranking']='count_keywords_mathes'

df['Main_category_names']= [classifier_subcateg.loc[value]['Cathegories'] for index, value in df['Main_category'].items()]
df['Pimary_subcategory']= [classifier_subcateg.loc[value]['Cathegories'] for index, value in df['1st_subcategory'].items()]
```

```python
subcat_dict
```

```python
import panel as pn
import plotly.express as px
from ipywidgets import widgets
pn.extension('plotly')
#subcat_dict1=subcat_dict
subcat_select = pn.widgets.Select(name='Select name of subcategory', options=subcat_dict, value='count_keywords_mathes')
func=subcat_select.value
@pn.depends(subcat_select.param.value)
def subcat_plot(func):
    temp_df = df.sort_values(by=[func], ascending=False).head(100)
    fig = px.scatter(temp_df, x="ecMaxContribution", y=func,
        size="durationM", color="Main_category_names",hover_data=['title','Main_category_names'],
        hover_name="acronym", log_x=True, size_max=40)
    fig.update_layout(legend=dict(orientation="h", yanchor="bottom",
        y=1.02, 
        xanchor="right",
        x=1), 
        yaxis={'title':'Ranking score'})
    fig.update_traces(hovertemplate="%{_other}: <br>Popularity: %{size}")
    return fig

dropdown=pn.Column(subcat_select,subcat_plot)
dropdown
```
